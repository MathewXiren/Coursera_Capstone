## Acquire Origin neighborhood data of Ottawa
Scrape the data from Wikipedia. Then, organize and clean the data.

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_K"
# IFrame(url, width=800, height=350)
df, = pd.read_html(url, match='Ottawa', skiprows=1)
# print(df)
xx = df.values.tolist()
# df.head()
# df = df[0].split()
# print(len(df))
ll=[]
for i in xx:
    for j in i:
        ll.append(j)
# print(ll)
df = pd.DataFrame(ll, columns=['Neighborhood'])
df.head()
```

```python
df["PostalCode"] = df.Neighborhood.str.extract('(\S\d\S)', expand=False)
```

```python
df.head()
```

```python
df["Borough"] = df.Neighborhood.str.extract('(\S\d\S\w+\()', expand=False)
df.head()
```

```python
df = df.dropna()
df['Neighborhood'] = df.Neighborhood.str.replace('(\S\d\S\w+\()', '')
df['Neighborhood'] = df.Neighborhood.str.replace(')', '')
df["Borough"] = df.Borough.str.replace('(', '')
df["Borough"] = df.Borough.str.replace('(\S\d\S)', '')
df["PostalCode"] = df.PostalCode.str.strip()
df.head()
```

```python
df=df[["PostalCode", 'Borough', 'Neighborhood']]
```

```python
df.head()
```

```python
# dff = df.groupby(['PostalCode', 'Borough']).agg(', '.join)
# dff.head()
```

```python
df = df[df.Borough == 'Ottawa']
df.reset_index
df.head()
```

```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
# clean the json and structure it into a pandas dataframe.
venues = results['response']['groups'][0]['items']
    
nearby_venues = json_normalize(venues) # flatten JSON

# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_venues =nearby_venues.loc[:, filtered_columns]

# filter the category for each row
nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)

# clean columns
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]

nearby_venues.head()
```

## Get venue information if multiple locations at once

```python
# for each location (postal code), explore venues
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

## Cluster
I will cluster the locations according to the distribution of the restaurant by using k-means. Then, we can further analyze type of each restaurant in a cluster, which will be convenient for us to determine the best restaurant category in a location.

```python
ottawa_restaurant.groupby('Neighborhood').count()
```

```python
print('There are {} uniques categories.'.format(len(ottawa_restaurant['Venue Category'].unique())))
```

```python
# # Run k-means to cluster the venues into 5 clusters.

# # set number of clusters
kclusters = 14

restaurant_grouped_clustering = ottawa_restaurant.drop(['Neighborhood', 'Neighborhood Latitude', 'Neighborhood Longitude', 'Venue', 'Venue Category'], 1)
restaurant_grouped_clustering.head()
```

```python
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(restaurant_grouped_clustering)
# check cluster labels generated for each row in the dataframe
kmeans.labels_
```

```python
restaurant_merged = ottawa_restaurant.copy()
restaurant_merged['Cluster Labels'] = kmeans.labels_
restaurant_merged.head()
restaurant_merged.shape
```

```python
# # create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i+x+(i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(restaurant_merged['Venue Latitude'], restaurant_merged['Venue Longitude'], restaurant_merged['Neighborhood'], restaurant_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```

```python
restaurant_onehot = pd.get_dummies(ottawa_restaurant[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
restaurant_onehot['Neighborhood'] = ottawa_restaurant['Neighborhood'] 
restaurant_onehot['Cluster Labels'] = kmeans.labels_
# move neighborhood column to the first column
fixed_columns = [restaurant_onehot.columns[-1]] + list(restaurant_onehot.columns[:-1])
restaurant_onehot = restaurant_onehot[fixed_columns]

restaurant_onehot.head()
restaurant_onehot.shape
# len(restaurant_onehot['Neighborhood'].unique())
restaurant_grouped = restaurant_onehot.groupby(['Cluster Labels','Neighborhood']).mean().reset_index()
```

```python
# sort the venues in descending order
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[2:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
#     print(row_categories_sorted)
    return row_categories_sorted.index.values[0:num_top_venues]

# create the new dataframe and display the top 10 venues for each neighborhood.
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Cluster Labels', 'Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Cluster Labels'] = restaurant_grouped['Cluster Labels']
neighborhoods_venues_sorted['Neighborhood'] = restaurant_grouped['Neighborhood']

for ind in np.arange(restaurant_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 2:] = return_most_common_venues(restaurant_grouped.iloc[ind, :], num_top_venues)
    
```
