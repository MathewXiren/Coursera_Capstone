```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
# IFrame(url, width=800, height=350)
df, = pd.read_html(url, match="Postal Code", skiprows=1)
df.columns = ["PostalCode", "Borough", "Neighborhood"]
df.head()
```

```python
# Only process the cells that have an assigned borough. Ignore cells with a borough that is Not assigned.
df = df[df.Borough != 'Not assigned']
df.head(10)
```

## Project Part 2: Getting the latitude and the longitude coordinates of each neighborhood.
1. Leveraging on the geocoder API.
2. Update the coordinates accordingly in the dataframe.
3. Display them out accordingly to the layout in the instruction.

```python
df['Latitude'] = ''
df['Longitude'] = ''

# loop through the postal code to find out the latitude and longitude
for index, data in df.iterrows():
    lat_lng_coords = None
    while(lat_lng_coords is None):
        g = geocoder.google('{}, Toronto, Ontario'.format(data['PostalCode']))
        print(g)
        lat_lng_coords = g.latlng
    data['Latitude'] = lat_lng_coords[0]
    data['Longitude'] = lat_lng_coords[1]
#     print ('PostalCode:', data['PostalCode'], 'Latitude:', data['Latitude'], 'Longitude:', data['Longitude'])
```
